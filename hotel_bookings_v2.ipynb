##### Create function total_nights

```python
hotel_bookings['total_nights'] = hotel_bookings.apply(lambda x: round(x['stays_in_weekend_nights'] + x['stays_in_week_nights'], 0), axis=1)
hotel_bookings.head()
```

##### Create another function grouping guests i.e. large group, family, couple, individual traveller

```python
def party_type(hotel_bookings):
    total_guests = hotel_bookings['total_guests']
    adults = hotel_bookings['adults']
    if total_guests > 6:
        return("Large Group")
    elif total_guests == 1:
        return("Individual Traveller")
    elif total_guests == 0:
        return("Unknown")
    else:
        if adults != total_guests:
            return("Family")
        else:
            if total_guests == 2:
                return("Couple")
            else: 
                return("Group")
```

```python
hotel_bookings["party_type"] = hotel_bookings.apply(party_type, axis=1)
```
